```python
class Demo1Dataset(torch.utils.data.Dataset):
    def __init__(self):
        data_pd=pd.DataFrame([range(100),[x*x for x in range(100)]]).T
        data_pd.columns=['x','y']

    def __len__(self):
        return len(data_pd)

    def __getitem__(self, idx):
        return data_pd.x.values[idx],data_pd.y.values[idx]
```

```python
dataloader=Demo1Dataset()
```

```python
for xx ,yy in dataloader:
    print(xx,yy)
```
