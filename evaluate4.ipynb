```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
nums = 100
```

```python
metrics = np.load('./results/dataset4_Autoformer_custom_ftM_sl17_ll17_pl7_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/metrics.npy')
pred = np.load('./results/dataset4_Autoformer_custom_ftM_sl17_ll17_pl7_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/pred.npy')  # 预测步长 + 特征
true = np.load('./results/dataset4_Autoformer_custom_ftM_sl17_ll17_pl7_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/true.npy')
```

```python
pred[:,:1,0].shape
```

```python
step = 6
```

```python
# 定义 Groundtruth 和 Predict 数据 第一个控制步长 第二个控制舵机
gt1 = true[:,:step,0].flatten()
pred1 = pred[:,:step,0].flatten()

plt.plot([x for x in gt1[:nums]], c='b',label='一号舵机真实值')
plt.plot([x for x in pred1[:nums]],c='g',label='一号舵机预测值')


# 添加图例和标签
plt.legend(loc='best',fontsize=12)
plt.xlabel('Step',size=15)
plt.ylabel('Value',size=15)
plt.show()
```
