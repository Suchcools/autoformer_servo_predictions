```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python

metrics=np.load('./results/creep_Autoformer_custom_ftM_sl7_ll7_pl1_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/metrics.npy')
pred=np.load('./results/creep_Autoformer_custom_ftM_sl7_ll7_pl1_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/pred.npy')
true=np.load('./results/creep_Autoformer_custom_ftM_sl7_ll7_pl1_dm512_nh8_el1_dl1_df2048_fc3_ebtimeF_dtTrue_Exp_0/true.npy')
# 定义 Groundtruth 和 Predict 数据
gt = np.squeeze(true, axis=1)
pred = np.squeeze(pred, axis=1)

plt.plot([x[0] for x in gt], c='b',label='一号舵机真实值')
plt.plot([x[0] for x in pred],c='g',label='一号舵机预测值')


# 添加图例和标签
plt.legend(loc='best',fontsize=12)
plt.xlabel('Step',size=15)
plt.ylabel('Value',size=15)
plt.show()
```
