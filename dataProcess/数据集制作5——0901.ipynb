```python
import pandas as pd
rawdata = pd.read_excel('./时序数据集10g_3.xlsx')
rawdata
```

```python
rawdata = rawdata[['时间戳','1号舵机','2号舵机','3号舵机','4号舵机','X','Y','Z']]
rawdata
```

```python
import pandas as pd
from datetime import datetime, timedelta


# 定义起始时间和时间间隔
start_date = datetime(2000, 1, 1)
time_interval = timedelta(minutes=10)

# 生成时间列
timestamps = [start_date + i * time_interval for i in range(len(rawdata))]
rawdata['时间戳'] = timestamps
rawdata.columns =['date','1','2','3','4','x','y','z']
rawdata = rawdata.dropna()
```

```python
rawdata.to_csv('../dataset/dataset3.csv',index=False,encoding='utf_8_sig')
```
