```python
import pandas as pd
rawdata = pd.read_excel('./时序数据集10g_3.xlsx')
rawdata
```

```python
# 指定要合并的行数n
n = 5  # 以3行为例，您可以根据需要进行调整

# 创建新的DataFrame来存储重构后的数据
new_data = {
    '时间戳': [],
    '1号舵机': [],
    '2号舵机': [],
    '3号舵机': [],
    '4号舵机': [],
    'X': [],
    'Y': [],
    'Z': [],
    '负载': []
}

# 添加额外的n列来存储下n行的X、Y、Z值
for i in range(1, n + 1):
    new_data[f'{i}X'] = []
    new_data[f'{i}Y'] = []
    new_data[f'{i}Z'] = []

# 遍历数据，将每一行与下n行的X、Y、Z值组合成新的数据
for i in range(len(rawdata) - n):
    row1 = rawdata.iloc[i]
    new_data['时间戳'].append(row1['时间戳'])
    new_data['1号舵机'].append(row1['1号舵机'])
    new_data['2号舵机'].append(row1['2号舵机'])
    new_data['3号舵机'].append(row1['3号舵机'])
    new_data['4号舵机'].append(row1['4号舵机'])
    new_data['X'].append(row1['X'])
    new_data['Y'].append(row1['Y'])
    new_data['Z'].append(row1['Z'])
    new_data['负载'].append(row1['负载'])
    
    # 遍历下n行数据并合并X、Y、Z值
    for j in range(1, n + 1):
        row_next = rawdata.iloc[i + j]
        new_data[f'{j}X'].append(row_next['X'])
        new_data[f'{j}Y'].append(row_next['X'])
        new_data[f'{j}Z'].append(row_next['X'])

# 创建新的DataFrame
rawdata = pd.DataFrame(new_data)

# 打印新的DataFrame
rawdata
```

```python
from datetime import datetime, timedelta


# 定义起始时间和时间间隔
start_date = datetime(2000, 1, 1)
time_interval = timedelta(minutes=10)

# 生成时间列
timestamps = [start_date + i * time_interval for i in range(len(rawdata))]
rawdata['时间戳'] = timestamps
rawdata = rawdata.dropna()
rawdata.rename(columns={'时间戳': 'date'}, inplace=True)
rawdata
```

```python
rawdata.to_csv('../dataset/dataset4.csv',index=False,encoding='utf_8_sig')
```
